<h1> Clean Datasets </h1>

<h3>Cleaning Traffic Counts</h3>

```python
# Drop unnecessary columns
cl_traffic = traffic_counts.drop(['id', 'count_date', 'file_no'], axis = 1)
```

```python
# Helper function to clean count columns
def clean_counts(df):
    
    # All count columns
    cols = [col for col in df if ('count' in col)]
    
    for col in cols:
        
        # Convert to numeric values, coerce and fill with 0 if error
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
    return df
```

```python
# Clean the traffic count columns, convert types
cl_traffic = clean_counts(cl_traffic)
cl_traffic.describe()
```

<h3> Cleaning Collisions </h3>

```python
# drop unnecessary columns
cl_collisions = collisions.drop('report_id', axis = 1)
```

```python
# Clean hit/run level, fill NaN with Neither
cl_collisions['hit_run_lvl'].fillna('NEITHER', inplace = True)
```

```python
# Clean date time, convert to datetime object
cl_collisions['date_time'] = cl_collisions['date_time'].apply(pd.to_datetime)
```

```python
# Clean null street types
# cl_collisions.loc[cl_collisions.street_type.isnull()]
cl_collisions['street_type'] = cl_collisions.street_type.fillna(" ")
```

## Testing Hypothesis: Beat 122 (PB) has higher average hurt (injured + killed) than other beats

```python
# Mean total hurt in PB, beat 122
observed_avg_hurt = pb_collisions.mean().loc['total_hurt']
observed_avg_hurt
```

```python
averages = []

for i in np.arange(50000):
    random_sample = cl_collisions.sample((pb_collisions.shape[0]), replace=False)
    new_average = np.mean(random_sample['total_hurt'])
    averages.append(new_average)
    
averages = np.array(averages)  
```

```python
pd.Series(averages, name='Average').plot(
    kind='hist', 
    bins=cl_collisions['police_beat'].nunique(),
    title='average hurt from randomly created collision sets'
)
plt.scatter(observed_avg_hurt, 0, color='red', s=40)
plt.show()
```

```python
p_value = np.count_nonzero(averages >= observed_avg_hurt) / 50000
p_value
```

# Merging Traffic and Collisions

```python
def convert_stype(stype):
    types = {
        'AV': 'AVENUE',
        'ST': 'STREET',
        'RD': 'ROAD',
        'DR': 'DRIVE',
        'BL': 'BOULEVARD',
        'PY': 'PARKWAY',
        'WY': 'WAY',
        'CT': 'COURT',
        'PL': 'PLACE'
    }
    return types.get(stype, " ")
```

```python
# Create a street_type column
cl_traffic['street_type'] = cl_traffic['street_name'].apply(lambda x: x[len(x) - 2:]).apply(convert_stype)
cl_traffic['street'] = cl_traffic['street_name'].apply(lambda x: x[:len(x) - 3])
cl_traffic['street_type'].value_counts()
```
